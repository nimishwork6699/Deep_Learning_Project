# For the callculation of cgp and the geeting placed

```python
x=torch.tensor(6.7)
y=torch.tensor(0.0)
```

```python
w=torch.tensor(1.0,requires_grad=True)
b=torch.tensor(0.0,requires_grad=True)
```

```python
w
```

```python
b
```

```python
z=w*x+b
z
```

```python
y_pred=torch.sigmoid(z)
y_pred
```

```python
import torch.nn as nn
```

```python
criterion= nn.BCELoss()
loss=criterion(y_pred,y)
loss
```

```python
loss.backward()
```

```python
w.grad
```

```python
b.grad
```
