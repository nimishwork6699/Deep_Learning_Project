```python
import torch
import torch.nn as nn

class Model(nn.Module): #it inherit the class from the modulle of nn

 def __init__(self,num_features):

  super().__init__()
  self.linear=nn.Linear(num_features,1)#(input,output)
  self.sigmoid=nn.Sigmoid()

 def forward(self,features):
  out=self.linear(features)
  out=self.sigmoid(out)

  return out
```

```python
# creating dataset
features=torch.rand(10,5)

# creating model
model=Model(features.shape[1]) #column wala

# call model for forward pass
 # we can do like
 # model.forward(features)--> but wont
model(features)
```

```python
# showing model weights
model.linear.weight
#model.linear.bias
```

# For HIDDEN LAYER

```python
class Model(nn.Module): #it inherit the class from the modulle of nn

 def __init__(self,num_features):

  super().__init__()
  self.linear1=nn.Linear(num_features,3)#(input,output)
  self.relu=nn.ReLU()
  self.linear2=nn.Linear(3,1)
  self.sigmoid=nn.Sigmoid()

 def forward(self,features):
  out=self.linear1(features)
  out=self.relu(out)
  out=self.linear2(out)
  out=self.sigmoid(out)

  return out
```

```python
# creating dataset
features=torch.rand(10,5)

# creating model
model=Model(features.shape[1]) #column wala

# call model for forward pass
 # we can do like
 # model.forward(features)--> but wont
model(features)
```

```python
from torchinfo import summary

summary(model,input_size=(10,5))
```

# USING SEQUENTIAL

```python
class Model(nn.Module): #it inherit the class from the modulle of nn

 def __init__(self,num_features):

  super().__init__()
  self.network=nn.Sequential(
  nn.Linear(num_features,3),#(input,output)
  nn.ReLU(),
  nn.Linear(3,1),
  nn.Sigmoid()
  )
 def forward(self,features):
  out=self.network(features)

  return out
```
